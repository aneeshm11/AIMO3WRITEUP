```python
class AIMO3Template:
    def get_system_content(self, prompt, tool_cfg):
        return SystemContent.new().with_model_identity(prompt).with_reasoning_effort(
            reasoning_effort=ReasoningEffort.HIGH).with_tools(tool_cfg)

    def apply_chat_template(self, sys_prompt, usr_prompt, tool_cfg):
        return [Message.from_role_and_content(Role.SYSTEM, self.get_system_content(sys_prompt, tool_cfg)),
                Message.from_role_and_content(Role.USER, usr_prompt)]
```

```python
class AIMO3Sandbox:
    _port_lock, _next_port = threading.Lock(), 50000

    @staticmethod
    def _port_is_free(port):
        with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
            s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
            try:
                s.bind(('127.0.0.1', port))
                return True
            except OSError:
                return False

    @classmethod
    def _get_next_ports(cls, count=5):
        with cls._port_lock:
            while True:
                candidates = list(range(cls._next_port, cls._next_port + count))
                cls._next_port += count
                if all(cls._port_is_free(p) for p in candidates):
                    return candidates

    def __init__(self, timeout, init_timeout=None):
        self._default_timeout = timeout
        self._owns_kernel = False
        self._client = None
        self._km = None

        ports = self._get_next_ports(5)
        env = os.environ.copy()

        env.update({'PYDEVD_DISABLE_FILE_VALIDATION': '1', 'PYDEVD_WARN_EVALUATION_TIMEOUT': '0',
                   'JUPYTER_PLATFORM_DIRS': '1', 'PYTHONWARNINGS': 'ignore', 'MPLBACKEND': 'Agg'})



        self._km = KernelManager()
        self._km.shell_port, self._km.iopub_port, self._km.stdin_port, \
            self._km.hb_port, self._km.control_port = ports
        self._km.start_kernel(env=env, extra_arguments=['--Application.log_level=CRITICAL'])
        self._client = self._km.blocking_client()
        self._client.start_channels()

        effective_init = init_timeout if init_timeout is not None else timeout
        self._client.wait_for_ready(timeout=effective_init)

        self._owns_kernel = True
        self.execute('import math, numpy, sympy, mpmath, itertools, collections\nmpmath.mp.dps = 64\n')

    def _format_error(self, tb):
        return ''.join(re.sub(r'\x1b\[[0-9;]*m', '', f) for f in tb
                      if 'File "' not in f or 'ipython-input' in f)

    def execute(self, code, timeout=None):
        effective_timeout = timeout or self._default_timeout
        msg_id = self._client.execute(code, store_history=True, allow_stdin=False, stop_on_error=False)
        stdout, stderr, start = [], [], time.time()
        while True:
            if time.time() - start > effective_timeout:
                self._km.interrupt_kernel()
                return f'[ERROR] Execution timed out after {effective_timeout} seconds'
            try:
                msg = self._client.get_iopub_msg(timeout=1.0)
            except queue.Empty:
                continue
            if msg.get('parent_header', {}).get('msg_id') != msg_id: continue
            mt, c = msg.get('msg_type'), msg.get('content', {})
            if mt == 'stream':
                (stdout if c.get('name') == 'stdout' else stderr).append(c.get('text', ''))
            elif mt == 'error':
                stderr.append(self._format_error(c.get('traceback', [])))
            elif mt in {'execute_result', 'display_data'}:
                if txt := c.get('data', {}).get('text/plain'):
                    stdout.append(txt if txt.endswith('\n') else f'{txt}\n')
            elif mt == 'status' and c.get('execution_state') == 'idle':
                break
        out, err = ''.join(stdout), ''.join(stderr)
        return f'{out.rstrip()}\n{err}' if err and out else (err or out or '[WARN] No output. Use print() to see results.')

    def close(self):
        with contextlib.suppress(Exception):
            if self._client: self._client.stop_channels()
        if self._owns_kernel and self._km:
            with contextlib.suppress(Exception): self._km.shutdown_kernel(now=True)
            with contextlib.suppress(Exception): self._km.cleanup_resources()

    def reset(self):
        self.execute('%reset -f\nimport math, numpy, sympy, mpmath, itertools, collections\nmpmath.mp.dps = 64\n')

    def __del__(self):
        self.close()
```

```python
class AIMO3Tool:
    def __init__(self, timeout, prompt, sandbox=None):
        self._local_jupyter_timeout, self._tool_prompt, self._jupyter_session = timeout, prompt, sandbox
        self._owns_session, self._execution_lock, self._init_lock = sandbox is None, threading.Lock(), threading.Lock()

    def _ensure_session(self):
        if self._jupyter_session is None:
            with self._init_lock:
                if self._jupyter_session is None:
                    self._jupyter_session = AIMO3Sandbox(timeout=self._local_jupyter_timeout)

    def _ensure_last_print(self, code):
        lines = code.strip().split('\n')
        if not lines: return code
        last = lines[-1].strip()
        if any(x in last for x in ['print', 'import']) or not last or last.startswith('#'): return code
        lines[-1] = 'print(' + last + ')'
        return '\n'.join(lines)

    @property
    def instruction(self): return self._tool_prompt

    @property
    def tool_config(self): return ToolNamespaceConfig(name='python', description=self.instruction, tools=[])

    def _make_response(self, output, channel=None):
        msg = Message(author=Author(role=Role.TOOL, name='python'),
                     content=[TextContent(text=output)]).with_recipient('assistant')
        return msg.with_channel(channel) if channel else msg

    def process_sync_plus(self, message):
        self._ensure_session()

        # final_script = self._ensure_last_print(message.content[0].text)

        final_script = message.content[0].text

        with self._execution_lock:
            try:
                output = self._jupyter_session.execute(final_script)
            except TimeoutError as exc:
                output = f'[ERROR] {exc}'
        return [self._make_response(output, channel=message.channel)]
```

```python
class AIMO3Solver:
    def __init__(self, cfg, port=8000):
        self.cfg, self.port = cfg, port
        self.base_url, self.api_key = f'http://0.0.0.0:{port}/v1', 'sk-local'
        self.template, self.encoding = AIMO3Template(), load_harmony_encoding(HarmonyEncodingName.HARMONY_GPT_OSS)
        self.stop_token_ids = self.encoding.stop_tokens_for_assistant_actions()
        self._preload_model_weights()
        self.server_process = self._start_server()
        self.client = OpenAI(base_url=self.base_url, api_key=self.api_key, timeout=self.cfg.session_timeout)
        self._wait_for_server()
        self._initialize_kernels()
        self.notebook_start_time, self.problems_remaining = time.time(), 50
        # Global quota across the full 50-problem run.
        # Tracks how many all-singleton (chaos) problems have been allowed into streams 6-8.
        self.chaos_problems_used = 0

    def _preload_model_weights(self):
        print(f'Loading model weights from {self.cfg.model_path} into OS Page Cache...')
        start, files, total = time.time(), [], 0
        for root, _, fnames in os.walk(self.cfg.model_path):
            for fn in fnames:
                fp = os.path.join(root, fn)
                if os.path.isfile(fp):
                    files.append(fp)
                    total += os.path.getsize(fp)
        with ThreadPoolExecutor(max_workers=self.cfg.workers) as ex:
            list(ex.map(lambda p: open(p, 'rb').read(), files))
        print(f'Processed {len(files)} files ({total/1e9:.2f} GB) in {time.time()-start:.2f} seconds.\n')

    def _start_server(self):
        cmd = [sys.executable, '-m', 'vllm.entrypoints.openai.api_server',
               '--seed', str(self.cfg.seed),
               '--model', self.cfg.model_path,
               '--served-model-name', self.cfg.served_model_name,
               '--tensor-parallel-size', '1',
               '--max-num-seqs', str(self.cfg.batch_size),
               '--gpu-memory-utilization', str(self.cfg.gpu_memory_utilization),
               '--host', '0.0.0.0',
               '--port', str(self.port),
               '--dtype', self.cfg.dtype,
               '--kv-cache-dtype', self.cfg.kv_cache_dtype,
               '--max-model-len', str(self.cfg.context_tokens),
               '--stream-interval', str(self.cfg.stream_interval),
               '--async-scheduling', '--disable-log-stats', '--enable-prefix-caching']
        self.log_file = open('vllm_server.log', 'w')
        return subprocess.Popen(cmd, stdout=self.log_file, stderr=subprocess.STDOUT,
                                start_new_session=True)

    def _wait_for_server(self):
        print('Waiting for vLLM server...')
        start = time.time()
        for _ in range(self.cfg.server_timeout):
            if (rc := self.server_process.poll()) is not None:
                self.log_file.flush()
                raise RuntimeError(
                    f'Server died with code {rc}. Full logs:\n{open("vllm_server.log").read()}\n')
            try:
                self.client.models.list()
                print(f'Server is ready (took {time.time()-start:.2f} seconds).\n')
                return
            except Exception:
                time.sleep(1)
        raise RuntimeError('Server failed to start (timeout).\n')

    def _initialize_kernels(self):
        print(f'Initializing {self.cfg.workers} persistent Jupyter kernels...')
        start = time.time()
        self.sandbox_pool = queue.Queue()
        created = 0
        total_attempts = 0
        max_total_attempts = self.cfg.workers * 2

        while created < self.cfg.workers and total_attempts < max_total_attempts:
            needed = self.cfg.workers - created
            with ThreadPoolExecutor(max_workers=needed) as ex:
                futures = [
                    ex.submit(lambda: AIMO3Sandbox(
                        timeout=self.cfg.jupyter_timeout,
                        init_timeout=self.cfg.init_kernel_timeout
                    ))
                    for _ in range(needed)
                ]
                for future in as_completed(futures):
                    total_attempts += 1
                    try:
                        self.sandbox_pool.put(future.result())
                        created += 1
                    except Exception as e:
                        print(f'[WARN] Kernel init failed (attempt {total_attempts}): ...')

        if created < self.cfg.workers:
            print(f'[WARN] Only {created}/{self.cfg.workers} kernels ready. Proceeding.')
        print(f'Kernels initialized in {time.time() - start:.2f} seconds.\n')


    def _scan_for_answer(self, text):

        patterns = [
            r'oxed\s*\{([^}]+)\}',
            r'\*\*\s*([\d,\s]+?)\s*\*\*',
        ]

        last_val, last_pos = None, -1

        for pattern in patterns:
            for m in re.finditer(pattern, text, re.IGNORECASE | re.DOTALL):
                raw = m.group(1)
                digits_only = re.sub(r'\D', '', raw)
                if not digits_only:
                    continue
                try:
                    val = int(digits_only)
                    if 0 <= val <= 99999 and m.start() > last_pos:
                        last_pos, last_val = m.start(), val
                except ValueError:
                    pass

        return last_val

    def _process_attempt(self, problem, sys_prompt, idx, stop_evt, deadline):
        start_time = time.time()
        if stop_evt.is_set() or time.time() > deadline:
            return {'Attempt': idx+1, 'Answer': None, 'Time Taken': 0,
                    'Python Calls': 0, 'Python Errors': 0, 'Response Length': 0}

        local_tool, sandbox, py_calls, py_errs, total_toks, ans = None, None, 0, 0, 0, None
        # seed = int.from_bytes(
        #     hashlib.blake2b(
        #         f"{self.cfg.seed}:{idx}".encode(),
        #         digest_size=4
        #     ).digest(),
        #     "big"
        # )

        seed = int(math.pow(self.cfg.seed + idx, 2))

        try:
            sandbox = self.sandbox_pool.get(timeout=self.cfg.sandbox_timeout)
            local_tool = AIMO3Tool(self.cfg.jupyter_timeout, self.cfg.tool_prompt, sandbox)
            conv = Conversation.from_messages(
                self.template.apply_chat_template(sys_prompt, problem, local_tool.tool_config))

            for _ in range(self.cfg.turns):
                if stop_evt.is_set() or time.time() > deadline:
                    break
                prompt_ids = self.encoding.render_conversation_for_completion(conv, Role.ASSISTANT)
                if (max_toks := self.cfg.context_tokens - len(prompt_ids)) < self.cfg.buffer_tokens:
                    break

                stream = self.client.completions.create(
                    model=self.cfg.served_model_name,
                    temperature=self.cfg.temperature,
                    max_tokens=max_toks,
                    prompt=prompt_ids,
                    seed=seed,
                    stream=True,
                    extra_body={
                        'min_p': self.cfg.min_p,
                        'stop_token_ids': self.stop_token_ids,
                        'return_token_ids': True
                    }
                )

                try:
                    tok_buf, txt_chunks = [], []
                    for chunk in stream:
                        if stop_evt.is_set() or time.time() > deadline:
                            break
                        if new_toks := chunk.choices[0].token_ids:
                            tok_buf.extend(new_toks)
                            total_toks += len(new_toks)
                            txt_chunks.append(chunk.choices[0].text)
                        if '}' in chunk.choices[0].text and (
                                ans := self._scan_for_answer(
                                    ''.join(txt_chunks[-self.cfg.search_tokens:]))):
                            # print(txt_chunks[-5000:])
                            break
                finally:
                    stream.close()

                if ans or not tok_buf:
                    break
                new_msgs = self.encoding.parse_messages_from_completion_tokens(
                    tok_buf, Role.ASSISTANT)
                conv.messages.extend(new_msgs)
                last = new_msgs[-1]

                if last.channel == 'final':
                    ans = self._scan_for_answer(last.content[0].text)
                    break
                if last.recipient == 'python':
                    py_calls += 1
                    resp = local_tool.process_sync_plus(last)
                    if any(x in (txt := resp[0].content[0].text)
                           for x in ['[ERROR]', 'Traceback', 'Error:']):
                        py_errs += 1
                    conv.messages.extend(resp)
        except Exception:
            py_errs += 1
        finally:
            if sandbox is not None:
                recovered = False
                try:
                    sandbox.reset()
                    self.sandbox_pool.put(sandbox)
                    recovered = True
                except Exception as reset_exc:
                    print(f'[WARN] Sandbox reset failed ({type(reset_exc).__name__}). Replacing kernel.')
                    try:
                        sandbox.close()
                    except Exception:
                        pass

                if not recovered:
                    # Replace the dead kernel so pool size stays constant
                    try:
                        replacement = AIMO3Sandbox(
                            timeout=self.cfg.jupyter_timeout,
                            init_timeout=self.cfg.init_kernel_timeout
                        )
                        self.sandbox_pool.put(replacement)
                        print('[INFO] Replacement kernel ready.')
                    except Exception as replace_exc:
                        print(f'[WARN] Replacement kernel failed: {replace_exc}. Pool shrunk by 1.')
        return {'Attempt': idx+1, 'Response Length': total_toks, 'Python Calls': py_calls,
                'Python Errors': py_errs, 'Time Taken': time.time() - start_time, 'Answer': ans}

    def _fallback_decision(self, valid_streams, all_results):
        """
        Best-effort decision when no clean rule fired — either a FuturesTimeoutError
        (frozen thread) or all futures returned without triggering any rule (safety net).
        """
        if not valid_streams:
            return 0

        counts      = Counter([s['ans'] for s in valid_streams]).most_common()
        valid_freqs = [c[1] for c in counts]
        na_count    = sum(1 for r in all_results if r['Answer'] is None)
        freqs       = sorted(valid_freqs + ([1] * na_count), reverse=True)
        lead_margin = freqs[0] - (freqs[1] if len(freqs) > 1 else 0) if freqs else 0

        if lead_margin >= 1:
            return counts[0][0]

        tied_candidates = [c[0] for c in counts if c[1] == valid_freqs[0]]
        best_ans, lowest_time = None, float('inf')
        for c_ans in tied_candidates:
            tot_time = sum(s['time'] for s in valid_streams if s['ans'] == c_ans)
            if tot_time < lowest_time:
                lowest_time, best_ans = tot_time, c_ans
        return best_ans

    def solve_problem(self, problem):
        print(f'\nProblem: {problem}\n')

        # user_input = f'{problem} {self.cfg.preference_prompt}'

        user_input = problem

        budget   = 15*60

#================================================


        # elapsed_global = time.time() - self.notebook_start_time
        # time_left      = self.cfg.notebook_limit - elapsed_global
        # reserved_time  = max(0, self.problems_remaining - 1) * self.cfg.base_problem_timeout

        # budget = time_left - reserved_time
        # budget = min(budget, self.cfg.high_problem_timeout)
        # budget = max(budget, self.cfg.base_problem_timeout)

#================================================

        deadline = time.time() + budget
        print(f'Hard Timeout Deadline: {budget:.2f} seconds\n')

        all_results         = []
        valid_streams       = []
        stop_evt            = threading.Event()
        final_chosen_answer = 0
        decided             = False

        # Per-problem chaos tracking.
        # chaos_problem:        True when this problem passed the stream-5 chaos gate.
        # chaos_stream5_answer: The last valid answer seen at the moment the chaos gate
        #                       fired. Used as fallback if all 8 streams still return
        #                       unique answers (stream-5 is the "sweet spot" — deep enough
        #                       to be rigorous, not yet derailed by token exhaustion).
        chaos_problem        = False
        chaos_stream5_answer = None

        def get_time_tiebreaker(tied_candidates):
            best_ans, lowest_time = None, float('inf')
            for c_ans in tied_candidates:
                tot_time = sum(s['time'] for s in valid_streams if s['ans'] == c_ans)
                if tot_time < lowest_time:
                    lowest_time, best_ans = tot_time, c_ans
            return best_ans

        executor = ThreadPoolExecutor(max_workers=self.cfg.attempts)
        try:
            futures = [
                executor.submit(self._process_attempt,
                                user_input, self.cfg.system_prompt, i, stop_evt, deadline)
                for i in range(self.cfg.attempts)
            ]

            try:
                remaining = max(1.0, deadline - time.time() - 2.0)

                for future in as_completed(futures, timeout=remaining):
                    try:
                        r = future.result()
                        all_results.append(r)

                        if r['Answer'] is not None:
                            valid_streams.append({'ans': r['Answer'], 'time': r['Time Taken']})

                        na_count       = sum(1 for res in all_results if res['Answer'] is None)
                        finished_count = len(valid_streams) + na_count

                        if finished_count < 3:
                            continue

                        counts      = Counter([s['ans'] for s in valid_streams]).most_common()
                        valid_freqs = [c[1] for c in counts]
                        freqs       = sorted(valid_freqs + ([1] * na_count), reverse=True)
                        lead_margin = (freqs[0] - (freqs[1] if len(freqs) > 1 else 0)
                                       if freqs else 0)

                        action, ans = "WAIT", None

                        # ── Rule 0: Cursed problem ────────────────────────────────────────
                        # 3+ NA streams: problem is pathologically hard or broken.
                        # Stop immediately, return best valid answer or 0.
                        if na_count >= 3:
                            action = "STOP"
                            ans    = counts[0][0] if counts else 0

                        # ── Rule 1: Hard cap (4 votes) ────────────────────────────────────
                        # Any candidate reaches 4 votes — unambiguous win.
                        # Only real answer votes count; NAs never trigger this.
                        elif valid_freqs and valid_freqs[0] >= 4:
                            action, ans = "STOP", counts[0][0]

                        # ── Rule 2: Clear leader (margin ≥ 2) ─────────────────────────────
                        # Leader is at least 2 votes ahead after counting NAs as singletons.
                        # Fires at any stream depth — this is the ES3-equivalent signal.
                        #
                        # Key cases covered:
                        #   [3]       stream 3  — unanimous (margin 3)
                        #   [3,1]     stream 4  — decisive (margin 2)  ← ES3 equivalent
                        #   [3,1,1]   stream 5  — decisive (margin 2)  ← ES3 equivalent
                        #   [3,1,1,1] stream 6  — still decisive
                        elif lead_margin >= 2:
                            action, ans = "STOP", counts[0][0]

                        # ── Rule 3: Chaos gate (exactly stream 5, all singletons) ──────────
                        # All 5 completed streams returned unique outcomes (valid or NA).
                        # No candidate has 2+ votes — a genuinely hard/chaotic problem.
                        #
                        # Quota available (< 5 chaos problems used this notebook run):
                        #   Tag as chaos, save stream-5 answer as sweet-spot fallback, WAIT.
                        #   The problem is allowed to continue into streams 6–8.
                        #
                        # Quota exhausted (≥ 5 chaos problems already used):
                        #   Stop immediately. Return the last arriving valid answer.
                        #   We cannot afford more 8-stream chaos burns this run.
                        elif finished_count == 5 and freqs == [1, 1, 1, 1, 1]:
                            if self.chaos_problems_used < 9:
                                self.chaos_problems_used += 1
                                chaos_problem        = True
                                chaos_stream5_answer = (valid_streams[-1]['ans']
                                                        if valid_streams else None)
                                action = "WAIT"
                            else:
                                action = "STOP"
                                ans    = valid_streams[-1]['ans'] if valid_streams else 0

                        # ── Rule 4: Forced end (all 8 streams exhausted) ─────────────────
                        # Full attempt budget consumed. Must pick a winner now.
                        #
                        #   A. No valid answers at all              → return 0.
                        #   B. Clear leader (margin ≥ 1)            → take leader.
                        #   C. Chaos problem, still all singletons  → stream-5 sweet-spot.
                        #      valid_freqs[0] == 1 means nobody ever got 2+ votes — the
                        #      model never converged. Stream-5 answer is the best single-
                        #      trajectory result before reasoning started to drift.
                        #   D. Exact tie among top candidates       → time tiebreaker.
                        elif finished_count == 8:
                            action = "STOP"
                            if not counts:
                                ans = 0
                            elif lead_margin >= 1:
                                ans = counts[0][0]
                            elif chaos_problem and valid_freqs and valid_freqs[0] == 1:
                                ans = chaos_stream5_answer if chaos_stream5_answer is not None else 0
                            else:
                                tied_candidates = [c[0] for c in counts
                                                   if c[1] == valid_freqs[0]]
                                ans = get_time_tiebreaker(tied_candidates)

                        if action == "STOP":
                            final_chosen_answer = ans if ans is not None else 0
                            decided = True
                            stop_evt.set()
                            for f in futures:
                                f.cancel()
                            break

                    except Exception as exc:
                        print(f'Future failed: {exc}')

            except FuturesTimeoutError:
                print('WARNING: as_completed timed out — one or more streams appear frozen. '
                      'Triggering fallback decision on completed results.')
                stop_evt.set()

        finally:
            # wait=False prevents blocking forever on a truly frozen thread.
            executor.shutdown(wait=False, cancel_futures=True)

        if not decided:
            print('Fallback decision triggered — using best available results.')
            final_chosen_answer = self._fallback_decision(valid_streams, all_results)

        self.problems_remaining = max(0, self.problems_remaining - 1)

        if all_results:
            df = pd.DataFrame(all_results)
            df['Answer']     = df['Answer'].astype('Int64')
            df['Time Taken'] = df['Time Taken'].round(2)
            display(df)

        print(f'\nFinal Answer Decided: {final_chosen_answer}\n')
        return final_chosen_answer if final_chosen_answer is not None else 0

    def __del__(self):
        if hasattr(self, 'server_process'):
            self.server_process.terminate()
            self.server_process.wait()
        if hasattr(self, 'log_file'):
            self.log_file.close()
        if hasattr(self, 'sandbox_pool'):
            while not self.sandbox_pool.empty():
                with contextlib.suppress(Exception):
                    self.sandbox_pool.get_nowait().close()
```

```python
def predict(id_: pl.DataFrame, question: pl.DataFrame, answer: Optional[pl.DataFrame] = None) -> pl.DataFrame:
    start_time = time.time()
    id_value  = id_.item(0)
    question_text = question.item(0)

    final_answer = 0
    max_retries  = 1
    last_exc     = None

    for attempt_num in range(max_retries + 1):
        try:
            gc.disable()
            final_answer = solver.solve_problem(question_text)
            gc.enable()
            gc.collect()
            last_exc = None
            break
        except Exception as exc:
            gc.enable()
            gc.collect()
            last_exc = exc
            print(f'[predict] Attempt {attempt_num + 1} raised {type(exc).__name__}: {exc}')
            if attempt_num < max_retries:
                print('[predict] Retrying solve_problem...')
                time.sleep(2)

    if last_exc is not None:
        print(f'[predict] All {max_retries + 1} attempts failed. Returning 0.')

    elapsed = time.time() - start_time
    print(f'Execution Time: {int(elapsed // 60)} min {elapsed % 60:.2f} sec')
    return pl.DataFrame({'id': id_value, 'answer': final_answer})
```

```python
import json
import time
from tqdm import tqdm

json_path = "./reference.json"

with open(json_path, "r") as f:
    data = json.load(f)

correct = 0
total = len(data)

for i in tqdm(range(total), desc="Evaluating"):
    sample = data[i]
    question_text = sample["question"]
    true_answer = str(sample["answer"]).strip()

    start = time.time()

    final_answer = solver.solve_problem(question_text)

    end = time.time()

    pred_answer = str(final_answer).strip()

    is_correct = pred_answer == true_answer
    if is_correct:
        correct += 1

    print(f"Predicted: {pred_answer}")
    print(f"True: {true_answer}")
    print(f"Correct: {is_correct}")
    print(f"Time taken: {end - start:.4f} sec")
    print(f"Time taken: {(end - start) / 60:.4f} min")


accuracy = correct / total
print("\n======================")
print(f"Final Accuracy: {accuracy * 100:.2f}% ({correct}/{total})")
```
